```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = cv2.imread('../Data/d7d4071d-f5f6-4eea-a2c2-a613e9315d17.webp')
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
##### CHANNEL ORDER 
# Matplotlib - RGB, Red Green Blue
# OPENCV - BGR Blue Green Red
#
# This is why the image looks weird when we plot in matplotlib. 
# Open cv reads in BGR order, matplotlib is expecting the RGB order.
```

```python
# Open CV can convert color spaces. Converting to Blue Green Red to Red Green Blue.
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# Reading images as greyscale.

img_grey = cv2.imread('../Data/d7d4071d-f5f6-4eea-a2c2-a613e9315d17.webp', cv2.IMREAD_GRAYSCALE)
```

```python
# Doesnt have a need for three color chanels. Since the image is grey scale.
img_grey.shape 
```

```python
plt.imshow(img_grey, cmap='gray')
```

```python
# Resizing images

fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (400, 100))

plt.imshow(new_img)
```

```python
# Can resize with ratio
w_ratio, h_ratio = 0.5, 0.5
new_img = cv2.resize(fix_img, (0, 0), fix_img, w_ratio, h_ratio)
plt.imshow(new_img)
```

```python
# Flipping images

new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
# Flipping images

new_img = cv2.flip(fix_img, 1)
plt.imshow(new_img)
```

```python
# Flipping images

new_img = cv2.flip(fix_img, -1)
plt.imshow(new_img)
```
